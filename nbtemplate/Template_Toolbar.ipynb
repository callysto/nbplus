<h3>Buttons</h3>

A format is predefined and applied. Each button comes with a predefined format. Format includes font size and alignment (center, left, right). 

<h4>Title Button</h4>

When pressed, this button creates a pre-formatted cell for the developer to edit. The cell acts like a markdown cell, except when the user press enter, the text is formatted for them. 

<h4>Subtitle Button </h4>

When pressed, this button creates a pre-formatted cell for the developer to edit. The cell acts like a markdown cell, except when the user press enter, the text is formatted for them. 

<h4>Question Button</h4>

When pressed, this button creates a pre-formatted cell for the developer to edit. The cell acts like a markdown, and tracks question number. 

<h4>Answer Button</h4>

When pressed this button displays a menu, and allows the developer to choose between multiple choice and open question. If the former, it allows the developer to choose number of possible answers, and to enter text for each one of them. Developer specifies what the correct answer is. If the latter, it displays a box for the user (a student for example) to enter their answer. 

<h4>Feedback Button</h4>

When pressed this button creates a pre-formatted cell for the developer to edit. The cell contents are feedback from exercises or questions. The cell acts like a markdown. 

<h3>Table of Contents</h3>

When pressed, this button identifies the title and subtitles within the notebook. A document navigation is built into the table of contents, such that when the user clicks on a subtitle they are directed into the corresponding area within the notebook.
