<h2 align='center'>Goal</h2>

I would like to make it as easy as possible for developers to format their notebooks. 

<h2 align='center'>What do I want to see in this notebook?</h2>

0. Title
1. Table of contents
2. Headers
3. Subheaders
4. Questions
5. Conclusion

This notebook is open for exploration - create!

<h2 align='center'>Idea</h2>

<h3>Buttons</h3>

A format is predefined and applied. Each button comes with a predefined format. Format includes font size and alignment (center, left, right). 

<h4>Insert Title Button</h4>

When pressed, this button prompts the developer to enter the title for their notebook. Two secondary buttons appear: and Edit Title Button and a Save Title Button. 

<h4>Insert Subtitle Button </h4>

When pressed, this button prompts the developer to enter subtibles for their notebook. Two secondary buttons appear: and Edit Subtitle Button and a Save Subtitle Button. 

<h4>Insert Multiple Choice Question Button</h4>

When pressed, this button prompts the user with two fillable boxes:

Box 1) Question box. This is the Question the developer wants to ask.

Box 2) Answer/feedback boxes. The user can add options as they see fit. 

Additionally, they can edit or save their question with an Edit Question Button or a Save Question Button.

<h4>Insert image</h4>

When pressed, this button prompts the developer to enter the path where the image is found. 

For example, if the path to the notebook is

/home/user/NotebookSampleA/mynotebook.ipynb

and the path to the image is

/home/user/NotebookSampleA/images/my_image.png

the developer enters "./images/my_image.png"

Secondary buttons appear to change or load the image. 

<h3>Table of Contents</h3>

Table of contents is built automatically as the user constructs their notebook.
